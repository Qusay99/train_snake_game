```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("training_protokoll.csv")
```

```python
df.describe()
```

```python
figure, axes = plt.subplots(nrows=2, ncols=1, figsize = (14,14))

axes[0].set_title("Hchst ereichter Fitnesswert vs. Generation")
axes[0].set_xlabel("Generation")
axes[0].set_ylabel("Wert der Fitnessfunktion")
axes[0].plot(df['gen_highest_Fitness'].values, c='#4424D6')

axes[1].set_title("Hchst erreichter Score vs. Generation")
axes[1].set_xlabel("Generation")
axes[1].set_ylabel("Hchst erreichter Score")
axes[1].plot(df['gen_higest_score'].values,c='#FB9902')

```
